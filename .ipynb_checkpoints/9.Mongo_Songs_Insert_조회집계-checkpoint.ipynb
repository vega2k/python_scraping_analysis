### Songs 정보를 저장하고 읽고 집계하기
1. data/songs.json 읽어서 mongodb에 저장한다.
2. mongodb에 연결해서 DB와 Collection을 생성한다
3. find(), aggregate() 함수를 사용해서 읽고, 집계를 해본다.

```python
!pip show pymongo
```

```python
import pymongo
import json

with open('data/songs.json','r',encoding='utf8') as file:
    contents = file.read()
    songs_json = json.loads(contents)

print(type(songs_json), len(songs_json))    
#songs_json[0:2]
```

```python
# 1. MongoDB와 연결
con = pymongo.MongoClient(host='localhost',port=27017)
print('Connection : ', con)
#2. Database 생성
db = con.song_db
print('Database : ', db)
print('DB Name : ', db.name)
#3. Collection 생성
col = db.songs
print('Collection : ',col)
print('Col Name : ', col.name)
```

```python
#songs_json 데이터를 insert_many() 함수 사용해서 document를 등록하기
col.insert_many(songs_json)

#document 건수 확인
col.count_documents({})
```

```python
#find() 함수 사용해서 document 읽기
print(type(col.find({})))
for doc in col.find({}):
    #print(type(doc))
    print(doc['가수'])
```

```python
#장르가 댄스인 노래의 가수,장르,발매일, _id는 빼고 가져오기
for song in col.find({'장르':'댄스'},{'_id':0,'가수':1,'장르':1,'발매일':1}):
    print(song)
```

```python
# criteria(검색조건), projection(필드)을 인자로 받아서 출력하는 함수
def print_song(criteria, projection):
    for idx,song in enumerate(col.find(criteria,projection),1):
        print(idx,song)
```

```python
#방탄소년단의 노래는?
criteria = {'가수':'방탄소년단'}
projection = {'_id':0,'가수':1,'곡명':1,'장르':1,'발매일':1}
print_song(criteria, projection)
```

```python
#방탄소년단 과 폴킴의 노래는?
criteria = {'가수':{'$in':['방탄소년단','폴킴']}}
print_song(criteria, projection)
```

```python
#앨범 이름 중에 OST 가 포함된 노래는?
criteria = {'앨범':{'$regex':'OST'}}
print_song(criteria, projection)
```

```python
def test(*args):
    agg_list = []
    for arg in args:
        agg_list.append(arg)
    print(agg_list)
    
    for song in col.aggregate(agg_list):
        print(song)
        
group = {
        '$group':{
            '_id':'$가수',
            '노래개수':{'$sum':1}
        }
    }
sort = {
        '$sort':{'노래개수':-1}
    }
test(group,sort)
```

```python
# 가수별로 노래가 몇 곡이나 있는지?

for song in col.aggregate([
    {
        '$group':{
            '_id':'$가수',
            '노래개수':{'$sum':1}
        }
    },
    {
        '$sort':{'노래개수':-1}
    },
    {
        '$project':{
            '가수이름':'$_id',
            '_id':0,
            '노래개수':1
        }
    }
]):
    print(song)
```

```python
# 장르별로 노래가 몇 곡이나 있는지?
group = {
   '$group':{
       '_id':'$장르',
       '노래개수':{'$sum':1}
   }     
}
sort = {
    '$sort':{'노래개수':-1}
}
project = {
    '$project':{
        '장르':'$_id',
        '_id':0,
        '노래개수':1
    }
}
for song in col.aggregate([group,sort,project]):
    print(song)
```
